```python
import pandas as pd


data = pd.read_csv('GSAF5.csv',encoding = "latin-1")
#data.head(10)
```

```python
#print(data.shape)
print(data.columns)
#print(data.dtypes)

data = data.drop('Unnamed: 22', 1)
data = data.drop('Unnamed: 23', 1)
data = data.drop('pdf', 1)
data = data.drop('href formula', 1)
data = data.drop('href', 1)
data = data.drop('Case Number.1', 1)
data = data.drop('Case Number.2', 1)
data = data.drop('original order', 1)
data = data.drop('Date', 1)
```
